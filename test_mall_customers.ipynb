```python
data = pd.read_csv("data/Mall_Customers.csv")
```

```python
data
```

```python
data = data.rename(columns=lambda x: re.sub(r'\W+', '', x))
```

```python
data
```

```python
label_col = ["Gender", "Age", "AnnualIncomek"]
```

```python
target_col = "SpendingScore1100"
```

```python
formula = str(target_col)+' ~ '+' + '.join(['%s' % variable for variable in label_col])
```

```python
with pm.Model():
    family = pm.glm.families.Normal()
    pm.GLM.from_formula(formula, data=data, family=family)
    trace = pm.sample(draws=1000, chains=2, tune=500, random_seed=23)
    pm.traceplot(trace)
```
